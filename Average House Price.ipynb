```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the data from CSV file
data = pd.read_csv(r'Average_house_price,_by_UK_country,_January_2005_to_July_2018.csv')


print(data.columns)
```

```python
# Remove spaces from column names
data.columns = data.columns.str.strip()
```

```python
##LINE GRAPH
def line_plot():
    """
    Generates a line plot of average house prices in the UK by country, from January 2005 to July 2018.
    """
   
    # Create a line plot with multiple lines
    plt.plot(data["Month"], data["England"], label="England")
    plt.plot(data["Month"], data["Wales"], label="Wales")
    plt.plot(data["Month"], data["Scotland"], label="Scotland")
    plt.plot(data["Month"], data["Northern Ireland"], label="Northern Ireland")

    # Set tick locations and labels for every 3 months
    tick_locations = range(0, len(data), 6)
    tick_labels = data.loc[tick_locations, "Month"]

    plt.xticks(tick_locations, tick_labels, rotation=90)

    # Add axis labels and a title
    plt.xlabel("Month")
    plt.ylabel("Average House Price")
    plt.title("Average House Prices by UK Country")

    # Add a legend
    plt.legend()

    # Display the plot
    plt.show()
```

```python
def stack_plot():
    
    """
    Plots a stacked area graph of the average house prices by UK country over time.

    Returns:
    None
    """
        
    # Define the data for each country
    england = data["England"]
    wales = data["Wales"]
    scotland = data["Scotland"]
    ni = data["Northern Ireland"]

    # Define the total prices for each month
    prices = [england[i] + wales[i] + scotland[i] + ni[i] for i in range(len(england))]

    # Create a stackplot with multiple lines
    plt.stackplot(data["Month"], england, wales, scotland, ni, labels=["England", "Wales", "Scotland", "Northern Ireland"])

    # Set tick locations and labels for every 3 months
    tick_locations = range(0, len(data), 6)
    tick_labels = data.loc[tick_locations, "Month"]

    plt.xticks(tick_locations, tick_labels, rotation=90)


    # Add axis labels and a title
    plt.xlabel("Month")
    plt.ylabel("Average House Price")
    plt.title("Average House Prices by UK Country")

    # Add a legend
    plt.legend()

    # Display the plot
    plt.show()
```

```python
def bar_plot():
    """
    Plots a bar graph of the latest average house prices by UK country.

    Returns:
    None
    """ 
    # Calculate the mean prices for each country
    england_mean = data["England"].mean()
    wales_mean = data["Wales"].mean()
    scotland_mean = data["Scotland"].mean()
    ni_mean = data["Northern Ireland"].mean()
    
    # Create a bar plot
    prices = [england_mean, wales_mean, scotland_mean, ni_mean]
    plt.bar(["England", "Wales", "Scotland", "Northern Ireland"], prices)
    
    # Add axis labels and a title
    plt.xlabel("Country")
    plt.ylabel("Average House Price")
    plt.title("Average House Prices by UK Country (2005-2018)")
    
    # Display the plot
    plt.show()
    
    """
    This function reads in the CSV file and calculates the average house prices for the whole duration of that dataset. 
    The function then creates a bar graph using the plt.bar function, passing in the country names and the corresponding prices. 
    The axis labels and title are added to the function and also setting the presentation of the x axis and then finally, the plot is displayed using plt.show().
    
    """
```
